# Netflix Clustering
Netflix has thousands of titles, making it difficult to explore content!  
Can we group these titles based on features like genre, rating and duration?   
Yes.  
By clustering shows/movie titles, we can uncover the hidden patterns that help in recommendations and content analysis.

## Objective
- Cluster titles using unsupervised ML algorithms - KMeans and DBSCAN 
- Visualize patterns across genres, ratings, and duration using t-SNE
  
## Dataset
- Columns: title, type, cast, release_year, rating, duration, etc,
- Number of records: 8808

```python
#import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans, DBSCAN
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#versions used
import sys, sklearn, matplotlib, seaborn, pandas, numpy
print("Python:", sys.version)
print("pandas:", pd.__version__)
print("sklearn:", sklearn.__version__)
print("matplotlib:", matplotlib.__version__)
```

```python
#load dataset
df = pd.read_csv('dataset/netflix_titles_set.csv')
df.head()
df.info()
df.describe()
df['type'].value_counts()
```

```python
#drop rows with missing values
df = df.dropna(subset=['director', 'cast', 'country', 'rating', 'duration'])
print(f"Shape after dropping rows with missing critical values: {df.shape}")
```

```python
#preprocess the duration
def preprocess_duration(duration):
    if 'min' in duration:
        return int(duration.split(' ')[0])
    elif 'Season' in duration:
        return int(duration.split(' ')[0]) * 60  
    return 0
df['duration'] = df['duration'].apply(preprocess_duration)
```

```python
#normalize numerical features
scaler = StandardScaler()
df[['release_year', 'duration']] = scaler.fit_transform(df[['release_year', 'duration']])
```

```python
#fill NaN values with empty strings
df['director'] = df['director'].fillna('')
df['cast'] = df['cast'].fillna('')
df['country'] = df['country'].fillna('')
```

```python
#combine text features for vectorization
df['text_features'] = df['type'] + ' ' + df['title'] + ' ' + df['director'] + ' ' + df['cast'] + ' ' + df['country'] + ' ' + df['rating'] + ' ' + df['listed_in'] + ' ' + df['description']
```

```python
#use tf-idf vectorizer to transform text features into numerical features
tfidf = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), max_df=0.9, min_df=5, max_features=1000)
tfidf_matrix = tfidf.fit_transform(df['text_features'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), index=df.index, columns=tfidf.get_feature_names_out())
```

```python
#combine numerical and text features
final_df = pd.concat([df[['release_year', 'duration']], tfidf_df], axis=1)
```

```python
#elbow method to find optimal no. of clusters 
sse = []
sil_scores = []
for k in range(2, 11):  
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(final_df)
    sse.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(final_df, kmeans.labels_))
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), sse, marker='o')
plt.title('Elbow Method For Optimal k', weight='bold')
plt.xlabel('Number of clusters', weight='bold')
plt.ylabel('SSE', weight='bold')
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), sil_scores, marker='o')
plt.title('Silhouette Score For Different k', weight='bold')
plt.xlabel('Number of clusters', weight='bold')
plt.ylabel('Silhouette Score', weight='bold')
plt.show()
```

```python
#apply K-means clustering
k = 5 
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(final_df)
df['cluster'] = kmeans.labels_
print(df.head())
score = silhouette_score(final_df, kmeans.labels_)
print(f'Silhouette Score: {score}')
```

```python
#pca for reducing dimensions - to visualize in 2D
pca = PCA(n_components=2, random_state=42)
pca_df = pca.fit_transform(final_df)
pca_df = pd.DataFrame(data=pca_df, columns=['PC1', 'PC2'])
pca_df['cluster'] = kmeans.labels_
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['cluster'], cmap='viridis', marker='o')
plt.title('Clusters of Netflix Movies and Shows', weight='bold')
plt.xlabel('Principal Component 1', weight='bold')
plt.ylabel('Principal Component 2', weight='bold')
plt.show()
```

```python
#apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5, n_jobs=-1)
dbscan_labels = dbscan.fit_predict(final_df)
df['cluster_dbscan'] = dbscan_labels
print(df.head())
unique_labels = set(dbscan_labels)
print(f'Number of clusters found by DBSCAN: {len(unique_labels) - (1 if -1 in dbscan_labels else 0)}')
```

```python
#visualize DBSCAN clusters using same PCA components
pca_df_dbscan = pd.DataFrame(data=pca_df, columns=['PC1', 'PC2'])
pca_df_dbscan['cluster'] = dbscan_labels
plt.figure(figsize=(10, 5))
plt.scatter(pca_df_dbscan['PC1'], pca_df_dbscan['PC2'], c=pca_df_dbscan['cluster'], cmap='viridis', marker='o')
plt.title('Clustering of netflix titles using DBSCAN', weight='bold')
plt.xlabel('Principal Component 1', weight='bold')
plt.ylabel('Principal Component 2', weight='bold')
plt.show()
```

```python
#use t-SNE for better visualization of clusters
tsne = TSNE(n_components=2, random_state=42, max_iter=300)
tsne_df = tsne.fit_transform(final_df)
tsne_df = pd.DataFrame(data=tsne_df, columns=['TSNE1', 'TSNE2'])
tsne_df['cluster'] = kmeans.labels_
plt.figure(figsize=(10, 5))
plt.scatter(tsne_df['TSNE1'], tsne_df['TSNE2'], c=tsne_df['cluster'], cmap='viridis', marker='o')
plt.title('Clustering of netflix titles using t-SNE', weight='bold')
plt.xlabel('t-SNE Component 1', weight='bold')
plt.ylabel('t-SNE Component 2', weight='bold')
plt.show()
```
